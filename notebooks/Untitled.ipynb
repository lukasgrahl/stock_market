```python
def create_df(list_etf, currency, eur_gbp, eur_usd):
    
    df = list_etf[0].history(period='max')['Close'].rename(list_etf[0].ticker)
    
    df = pd.DataFrame(df)
    
    df = pd.concat([df, eur_gbp, eur_usd], axis=1)
    
    list_etf = list_etf[1:]
    
    for etf in list_etf:
        
        etf_col = pd.DataFrame(etf.history(period='max')['Close'].rename(etf.ticker))
        
        df = pd.merge(df,
                     etf_col,
                     left_index=True,
                     right_index=True,
                     how='outer')
        
        if currency[etf.ticker]  == 'eur_usd':
            
            df[etf.ticker] = df[etf.ticker] * df['eur_usd']
            
        elif currency[etf.ticker] == 'eur_gbp':
            
            df[etf.ticker] = df[etf.ticker] * df['eur_gbp']
            
        else:
            
            df[etf.ticker] = df[etf.ticker]

    return df
```

```python
def insert_inflows(df_in, inflows):
    
    df_out = pd.DataFrame(data=df_in.index).set_index('Date')
    
    #df_in = df_in.drop(['eur_gbp', 'eur_usd'], axis=1)
    
    for item in inflows:
    
        inflow_date = pd.Timestamp(datetime.datetime.strptime(item[0], '%d/%m/%Y').date())

        work_df = pd.concat([df_in,
                            (df_in.loc[inflow_date:].apply(
                                lambda x: x / x.dropna()[0]).apply(
                                lambda x: x * weights[x.name]).sum(axis=1) * item[1]).rename(f'inflow_{item[0]}')],
                           axis=1)
        
        new_col = pd.Series(work_df[f'inflow_{item[0]}'])
        
        df_out = pd.concat([df_out, new_col], axis=1)
    
    return df_out
```
